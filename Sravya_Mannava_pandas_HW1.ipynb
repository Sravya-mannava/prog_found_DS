### Task 11
Create a new column titled `CS5000 Grade` in the score DataFrame to calcualte the letter grade for each student in course CS 5000 using the following criteria
+ `A: score >= 90`
+ `B: 80 <= score < 90`
+ `C: 70 <= score < 80`
+ `D: 60 <= score < 70`
+ `F: score < 60`

```python
conditions = [ (scores['CS 5000'] >= 90),
               (scores['CS 5000'] < 90) & (scores['CS 5000'] >=80),
               (scores['CS 5000'] < 80) & (scores['CS 5000'] >=70),
               (scores['CS 5000'] < 70) & (scores['CS 5000'] >=60),
               (scores['CS 5000'] < 60)]
results = ['A','B','C','D','F']
scores['CS5000 Grade'] = np.select(conditions,results)
scores
```

### Task 13
Using three courses' scores (assuming each course has 3 credits), calculate each student's GPA for Spring 2023, respectively. The following criteria should be used in the calculation
+ `A: 4.0`
+ `B: 3.0`
+ `C: 2.0`
+ `D: 1.0`
+ `F: 0.0`

For example, if a student's grades are A, B and C. That student's GPA should be calculated as

`(3 * 4 + 3 * 3 + 3 * 2) / 9 = 3.0`

Add a new column titled `GPA` in the scores DataFrame to store the GPA for each student

```python
conditions_1 = [ (scores['CS 5000'] >= 90),
               (scores['CS 5000'] < 90) & (scores['CS 5000'] >=80),
               (scores['CS 5000'] < 80) & (scores['CS 5000'] >=70),
               (scores['CS 5000'] < 70) & (scores['CS 5000'] >=60),
               (scores['CS 5000'] < 60)]
conditions_2 = [ (scores['CS 5110'] >= 90),
               (scores['CS 5110'] < 90) & (scores['CS 5110'] >=80),
               (scores['CS 5110'] < 80) & (scores['CS 5110'] >=70),
               (scores['CS 5110'] < 70) & (scores['CS 5110'] >=60),
               (scores['CS 5110'] < 60)]
conditions_3 = [ (scores['CS 5600'] >= 90),
               (scores['CS 5600'] < 90) & (scores['CS 5600'] >=80),
               (scores['CS 5600'] < 80) & (scores['CS 5600'] >=70),
               (scores['CS 5600'] < 70) & (scores['CS 5600'] >=60),
               (scores['CS 5600'] < 60)]
result = [4.0,3.0,2.0,1.0,0.0]
scores["GPA"] =  ((np.select(conditions_1,result)* 3)  + (np.select(conditions_2,result)* 3) + (np.select(conditions_3,result)* 3)) / 9
scores
```

### Task 14
Insert a new column title `Status` to provide a student academic standing based on the GPA
+ `Honor Roll: GPA >= 3.5`
